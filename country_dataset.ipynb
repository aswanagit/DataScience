```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("data_csv.csv")
plt.figure(figsize=(10,3))
x=data["Country Code"]
plt.hist(x,bins=30,color="red")
plt.title("birds")
plt.xlabel("Country Code")
plt.ylabel("count")
plt.show
```
